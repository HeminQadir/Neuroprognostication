```python
import torch
import torch.fft as fft
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Check if a GPU is available
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")
```

```python
# Parameters for the synthetic signal
sampling_rate = 1000  # Hz
duration = 10  # seconds
num_samples = sampling_rate * duration
time = torch.arange(0, duration, 1/sampling_rate, device=device)
frequency = 10  # Frequency of the signal in Hz
amplitude = 1.0
noise_amplitude = 0.2
```

```python
# Create a synthetic EEG-like signal with noise on the GPU
eeg_signal = amplitude * torch.sin(2 * np.pi * frequency * time) + noise_amplitude * torch.randn(num_samples, device=device)
```

```python
# Calculate the FFT of the EEG-like signal
eeg_fft = fft.fft(eeg_signal)
```

```python
# Define the bandpass filter in the frequency domain

low_cutoff = 1  # Lower cutoff frequency in Hz
high_cutoff = 50  # Upper cutoff frequency in Hz
```

```python
freqs = fft.fftfreq(eeg_signal.size(-1), 1.0 / sampling_rate)
bandpass_filter = (low_cutoff <= freqs) & (freqs <= high_cutoff)
```

```python
# Apply the bandpass filter to the frequency domain data
eeg_fft_filtered = eeg_fft * bandpass_filter.to(device, dtype=eeg_fft.dtype)
```

```python
eeg_filtered = fft.ifft(eeg_fft_filtered)
```

```python
eeg_filtered = eeg_filtered.to('cpu')

eeg_filtered_numpy = np.asarray(eeg_filtered , dtype=np.float64)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(time.to("cpu").numpy(), eeg_filtered_numpy, label='Filtered EEG Signal', color='orange')
plt.title("Original EEG-Like Signal")
plt.xlabel("Time (s)")





# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
# Calculate the frequency values corresponding to the FFT components on the CPU
freqs = np.fft.fftfreq(num_samples, 1.0 / sampling_rate)
freqs = torch.tensor(freqs, device=device)

mask = (freqs >= low_cutoff) & (freqs <= high_cutoff)
```

```python
# Apply the filter
eeg_fft_filtered = eeg_fft * mask
```

```python
# Calculate the inverse FFT to convert the filtered signal back to the time domain
eeg_filtered = fft.ifft(eeg_fft_filtered)
```

```python
# Plot the original and filtered EEG-like signals in the time domain
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(time.to("cpu").numpy(), eeg_signal.to("cpu").numpy())
plt.title("Original EEG-Like Signal")
plt.xlabel("Time (s)")
```

```python
plt.subplot(2, 1, 2)
plt.plot(time.to("cpu").numpy(), eeg_filtered.to("cpu").numpy().real)
plt.title("Filtered EEG-Like Signal")
plt.xlabel("Time (s)")

plt.tight_layout()
plt.show()
```
